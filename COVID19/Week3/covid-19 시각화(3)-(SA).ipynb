```python
def day_count(df, col, new_name):
    l = list()
    for i in df.index:
        if (i==0):
            l.append(df.loc[i,col])
        else:
            l.append(df.loc[i, col] - df.loc[i-1, col])
    df[new_name] = l
```

```python
def square2(num):
    for i in range(20,-1,-1):
        if num>=2**i:
            s=(i+(num-2**i)/2**i)
            break
        else:
            s=-1
    return s
    
        
```

전체적으로 신규확진자 수가 2^4에서 한번 꺾이면  금방 급감 그래프를 그린다. 거의 누적 확진자수가 2^5가 되기전에 신규확진자수가 거의 없는 모습

정리

초반 -1~5초반부 까지는 아무 의미가 없는 것같다. 몇몇 도시는 올라갔다 내려오는 그래프를 보여주기도 하지만 몇몇 도시는 선형성이 보이지 않았다.눈여결 볼 만한 것은 2^5+2^4, 즉 누적 확진자가 48명 아래일 때 감소하는 그래프가 그려지면 추가확진자는 거의 0이 대부분이다.초기 대응이 상당히 중요한 것 같다.

48명을 넘어서도 증가 그래프를 그린 곳은 총 7곳이다. 대구,경북,서울,경기도,충남,부산,경남 정도가 있는데 이중 경북 대구는 폭발적으로 증가 하다 급감하는 그래프를 그리고 있고 충남,부산,경남,서울,경기도 모두 2^6~2^7에서 꺾이는데 서울과 경기도는 다시 신규확진자 수가 증가한다. 대구경북 처럼 급 증가 하는 것은 아니고 2^4 정도를 유지하는 정도이다.아마 어떤 사건이 있지가 않고서는 적정선을 유지하면서 줄어들 것 같다.

많은 그래프 들이 한번에 많이 뛰고 훅 가라앉는 경향을 보이는데 아마 표본이 적어서 그럴것이다. 0,1,2차이와 4,5,6 차이가 매우 다르기 때문에 조심해서 그래프를 봐야할 것 같다.
